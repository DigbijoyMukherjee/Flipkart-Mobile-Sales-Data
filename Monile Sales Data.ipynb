## Exploratory Analysis and Visualization




Let's begin by importing`matplotlib.pyplot` and `seaborn`.

```python
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (9, 5)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

Plotted a graph of pairplot of brand to understand the pricing and rating of different brands.
It can be understood in below obtained graph that samsung, apple and oppo are brands that have relatively higher rating and 
and most of the brand price range lies between 15000 to 30000 most.

```python
sns.set_style("white")
```

```python
sns.pairplot(fkt_df, hue='Brand');
```

Also plotted below displot to show how count of selling price and original price the motive to show this plotting is to understand the market saturation of brands on relation to pricing and finally after seeing the bar plot we can conclude, "It is quite clear that the market is saturated with lower to mid range cost phones i.e. <25000 INR"

```python
sns.displot(fkt_df, x='Selling_Price',bins=[5000,10000,15000,20000,25000,30000,35000,40000,50000,60000,80000], aspect=1.2,color='#fd6c9e')
```

```python
sns.displot(fkt_df, x='Original_Price',bins=[5000,10000,15000,20000,25000,30000,35000,40000,50000,60000,80000], aspect=1.2,color='#ff8243')
```

# So this can be said that the market is saturated with lower to mid range cost phones i.e. <25000 INR"
